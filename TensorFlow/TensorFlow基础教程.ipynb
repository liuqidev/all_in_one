**TensorFlow** is an open source software library for numerical computation using
data flow graphs.  The graph nodes represent mathematical operations, while
the graph edges represent the multidimensional data arrays (tensors) that flow
between them.  This flexible architecture enables you to deploy computation to one
or more CPUs or GPUs in a desktop, server, or mobile device without rewriting
code.  TensorFlow also includes [TensorBoard](https://www.tensorflow.org/guide/summaries_and_tensorboard), a data visualization toolkit.

TensorFlow was originally developed by researchers and engineers
working on the Google Brain team within Google's Machine Intelligence Research
organization for the purposes of conducting machine learning and deep neural
networks research.  The system is general enough to be applicable in a wide
variety of other domains, as well.

## 举例

$$
h=ReLU(Wx+b)
$$

**Variables** are stateful nodes which output their current value.
State is retained across multiple excutions of a graph.(mostly parameters)
反映在计算图上是`W`&`b`。

**Placeholders** are nodes whose value is fed in at excution time.(inputs, labels, ...)
如`x`

**Mathematical operations**
MatMul:Multiply two matrix values
Add:Add elementwise(with broadcasting)
ReLu:Activate with elementwise rectified linear function.
...


**Fetches:** List of graph nodes. Return the outputs of these nodes.

**Feeds:** Dictionary mapping from graph nodes to concrete value. Specifies the value of each graph node given in the dictionary.


Getting output

sesss.run(fetches, feeds)

1. Build a graph
    - Feedforwad/Prediction
    - Optimization(gradients and train_step operation)
2. Initialize a session
3. Train with session.run(train_step, feed_dict)

## 举例：线性回归（Linear Regression）


````python
x_batch = np.linspace(-1, 1, 101)
a = 2.0
b = 0.3
y_batch = a*x_batch + np.random.randn(*x_batch.shape)*b

fig = plt.figure(figsize=(8,6),dpi=72)
plt.scatter(x_batch, y_batch, color='green')
plt.show()
````

````python
# cs224：linear regression 
# Author: xxx
# Apdated from https://github.com/nlintz/TensorFlow-Tutorials

# 适配py2和py3
from __future__ import absolute_import 
from __future__ import print_function
from __future__ import division

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf
# matplotlib.use('TKAgg')

def create_data():
    x_batch = np.linspace(-1, 1, 101)
    a = 2.0
    b = 0.3
    y_batch = a*x_batch + np.random.randn(*x_batch.shape)*b
    return x_batch, y_batch

def linear_regression():
    x = tf.placeholder(tf.float32, shape=(None,), name='x')
    y = tf.placeholder(tf.float32, shape=(None,), name='y')
    
    with tf.variable_scope('lreg') as scope:
        w = tf.Variable(np.random.normal(), name='w')
        y_pred = tf.multiply(w, x)
        
        loss = tf.reduce_mean(tf.square(y_pred-y)) # 最重要！
        
    return x, y, y_pred, loss, w

def run():
    # 1.提供原始数据
    x_batch, y_batch = create_data()
    # 2.创建计算图结构
    x, y, y_pred, loss, W = linear_regression()
    # 3.运行
    optimizer = tf.train.GradientDescentOptimizer(0.1).minimize(loss)
    
    init = tf.global_variables_initializer()
    with tf.Session() as session:
        session.run(init)
        
        feed_dict = {x: x_batch, y: y_batch}
        
        for i in range(50):
            loss_val, _ = session.run([loss, optimizer], feed_dict)
            print('Iteration ', i, ' loss: ', loss_val.mean(), 'k: ', session.run(W))
            
        y_pred_batch = session.run(y_pred, {x: x_batch})
    
    plt.figure(figsize=(8, 6), dpi=120)
    plt.scatter(x_batch, y_batch)
    plt.scatter(x_batch, y_pred_batch, color='green', linewidth=0.5)
    plt.savefig('plot.png')
    plt.show()
    
if __name__=='__main__':
    run()
````

````python
# 另一个linear regression 
# Author: 莫烦Python
# https://www.bilibili.com/video/av16001891/?p=9

from __future__ import absolute_import 
from __future__ import print_function
from __future__ import division

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf

# create data
def generate_data():
    x = np.random.randn(100).astype(np.float32)
    y = x*0.1+0.3
    return x, y


x_data, y_data = generate_data()

# 创建计算图结构--开始
W = tf.Variable(tf.random_uniform([1], -0.1, 1.0))
b = tf.Variable(tf.zeros([1]))

y = W*x_data+b

loss = tf.reduce_mean(tf.square(y-y_data))
optimizer = tf.train.GradientDescentOptimizer(0.5)
train = optimizer.minimize(loss)

init = tf.initialize_all_variables()
# 创建计算图结构--结束

sess = tf.Session()
sess.run(init) # 非常重要！！！

for step in range(201):
    sess.run(train)
    
    if step%20 ==0:
        print(step, sess.run(W), sess.run(b))
````

````python
from __future__ import absolute_import 
from __future__ import print_function
from __future__ import division

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf

def add_layer(inputs, in_size, out_size, activation_function=None):
    """
    Add one more layer and return the output of this layer.
    """
    W = tf.Variable(tf.random_normal([in_size, out_size]))
    b = tf.Variable(tf.zeros([1, out_size])+0.1)
    Wx_plus_b = tf.matmul(inputs, W)+b
    if activation_function == None:
        outputs = Wx_plus_b
    else:
        outputs = activation_function(Wx_plus_b)
    return outputs

# load the data
x_data = np.linspace(-1, 1, 300, dtype=np.float32)[:, np.newaxis]
# print(x_data.shape) # (300, 1) 
# add some noisy
noise = np.random.normal(0, 0.05, x_data.shape).astype(np.float32)
# print(noise.shape) # (300, 1)
y_data = np.square(x_data)-0.5 + noise


# defie placeholder for inputs to network
xs = tf.placeholder(tf.float32, [None, 1])
ys = tf.placeholder(tf.float32, [None, 1])

# add hidden layer
l1 = add_layer(xs, 1, 10, activation_function=tf.nn.relu)
# add output layer
y_pred = add_layer(l1, 10, 1,activation_function=None)

# 计算预测值和真实值之间的误差
loss =tf.reduce_mean(tf.reduce_sum(tf.square(y_data-y_pred),
                                  reduction_indices=[1]
                                  ))

train_step = tf.train.GradientDescentOptimizer(0.1).minimize(loss)

# 初始化变量
# init = tf.initialize_all_verables() # 过时
init = tf.global_variables_initializer()

with tf.Session() as sess:
    # initialize
    sess.run(init)
    
    feed_dict = {xs: x_data, ys:y_data}
    for i in range(1000):
        # training
        sess.run(train_step, feed_dict=feed_dict) # 注意这句话的位置
        if i%50 ==0:
            print('Loss: ', sess.run(loss, feed_dict={xs:x_data, ys: y_data}))
    
````

## Optimizer 优化器

## TensorBoard 


````python
from __future__ import absolute_import 
from __future__ import print_function
from __future__ import division

import numpy as np
import matplotlib
import matplotlib.pyplot as plt
import tensorflow as tf

def add_layer(inputs, in_size, out_size, n_layer, activation_function=None):
    """
    Add one more layer and return the output of this layer.
    """
    layer_name = 'layer%s'%n_layer
    with tf.name_scope('layer'):
        with tf.name_scope('weights'):
            W = tf.Variable(tf.random_normal([in_size, out_size]), name='W')
            tf.summary.histogram(layer_name+'/weights', W)
        with tf.name_scope('biases'):
            b = tf.Variable(tf.zeros([1, out_size])+0.1, name='b')
            tf.summary.histogram(layer_name+'/biases', b)

        with tf.name_scope('Wx_plus_b'):
            Wx_plus_b = tf.matmul(inputs, W)+b
            tf.summary.histogram(layer_name+'/Wx_plus_b', Wx_plus_b)
        if activation_function == None:
            outputs = Wx_plus_b
        else:
            outputs = activation_function(Wx_plus_b)
        tf.summary.histogram(layer_name+'/outputs', outputs)
        return outputs

# load the data
x_data = np.linspace(-1, 1, 300, dtype=np.float32)[:, np.newaxis]
# print(x_data.shape) # (300, 1) 
# add some noisy
noise = np.random.normal(0, 0.05, x_data.shape).astype(np.float32)
# print(noise.shape) # (300, 1)
y_data = np.square(x_data)-0.5 + noise

with tf.name_scope('outputs'):
    # define placeholder for inputs to network
    xs = tf.placeholder(tf.float32, [None, 1], name='x_in')
    ys = tf.placeholder(tf.float32, [None, 1], name='y_out')

# add hidden layer
l1 = add_layer(xs, 1, 10,n_layer=1, activation_function=tf.nn.relu)
# add output layer
y_pred = add_layer(l1, 10, 1,n_layer=2, activation_function=None)

with tf.name_scope('loss'):
# 计算预测值和真实值之间的误差
    loss =tf.reduce_mean(tf.reduce_sum(tf.square(y_data-y_pred),
                                      reduction_indices=[1],
                                        
                                      ))
    tf.summary.scalar('loss', loss)
    
with tf.name_scope('train'):
    train_step = tf.train.GradientDescentOptimizer(0.1).minimize(loss)

with tf.Session() as sess:
    merged = tf.summary.merge_all()
    #writer = tf.summary.FileWriter('logs/', sess.graph)
    writer = tf.summary.FileWriter('logs/', sess.graph)
    # 初始化变量
    # init = tf.initialize_all_verables() # 过时
    init = tf.global_variables_initializer()
    # initialize
    sess.run(init)
    
    feed_dict = {xs: x_data, ys:y_data}
    for i in range(1000):
        # training
        sess.run(train_step, feed_dict=feed_dict) # 注意这句话的位置
        if i%50 ==0:
            print('Loss: ', sess.run(loss, feed_dict={xs:x_data, ys: y_data}))
            result = sess.run(merged, feed_dict={xs: x_data, ys:y_data})
            writer.add_summary(result, i)
````
